```python
import cv2
from google.colab.patches import cv2_imshow

img_path = 'data/lena.png'

img = cv2.imread(img_path)
img_hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)

cv2_imshow(img)
cv2_imshow(img_hls)
```

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread(img_path)
img_lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)

cv2_imshow(img)
cv2_imshow(img_lab)
```
